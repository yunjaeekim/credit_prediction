```python
import os
from pdf2image import convert_from_path

def process_multiple_pdfs(pdf_folder, output_base_folder, pop_path):
    """
    지정된 폴더 내의 모든 PDF 파일을 변환하여 각 PDF 이름과 동일한 폴더에 JPG 파일로 저장.
    
    Args:
        pdf_folder: PDF 파일들이 있는 폴더 경로
        output_base_folder: 변환된 이미지를 저장할 기본 폴더 경로
    """
    # PDF 폴더 내 모든 파일 가져오기
    pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]
    
    if not pdf_files:
        print("No PDF files found in the folder.")
        return
    
    # 모든 PDF 파일 처리
    for cnt,pdf_file in enumerate(pdf_files):
        print(cnt,'번째 실행 중 / 전체 ', len(pdf_files))
        pdf_path = os.path.join(pdf_folder, pdf_file)
        
        # PDF 파일 이름을 따서 폴더 생성 후 이미지 저장
        pdf_name = os.path.splitext(pdf_file)[0]
        output_folder = os.path.join(output_base_folder, pdf_name)
        os.makedirs(output_folder, exist_ok=True)
        
        # PDF를 이미지로 변환
        images = convert_from_path(pdf_path, poppler_path= pop_path)
        
        # 각 페이지를 JPG로 저장
        for i, image in enumerate(images):
            jpg_filename = os.path.join(output_folder, f'page_{i+1}.jpg')
            image.save(jpg_filename, 'JPEG')
            print(f"Saved: {jpg_filename}")

# 사용 예시
pop_path = 'C:/Users/kyj93/OneDrive/바탕 화면/딥러닝_프로젝트/bin'
pdf_folder = 'C:/Users/kyj93/OneDrive/바탕 화면/딥러닝_프로젝트/2020_pdf'  # PDF 파일들이 있는 폴더 경로
output_base_folder = 'C:/Users/kyj93/OneDrive/바탕 화면/딥러닝_프로젝트/20_jpg'  # 변환된 이미지를 저장할 기본 폴더 경로

# 변환 실행
process_multiple_pdfs(pdf_folder, output_base_folder, pop_path)
```
